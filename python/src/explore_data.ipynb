# Loading in the data

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../data/FPA_FOD_20170508.sqlite')

query_1 = """
        SELECT
            SOURCE_SYSTEM_TYPE, NWCG_REPORTING_AGENCY, FIRE_CODE, FIRE_NAME, FIRE_YEAR,
            datetime(DISCOVERY_DATE) AS DISCOVERY_DATE, DISCOVERY_DOY, DISCOVERY_TIME, STAT_CAUSE_CODE,
            STAT_CAUSE_DESCR, datetime(CONT_DATE) AS CONT_DATE, CONT_DOY, CONT_TIME, FIRE_SIZE, FIRE_SIZE_CLASS,
            LATITUDE, LONGITUDE, STATE, COUNTY, FIPS_CODE, FIPS_NAME
        FROM
            Fires
        """

df_1 = pd.read_sql_query(query_1, conn)
```

# Exploratory data analysis

## How many observations have NULL values at FIRE_CODE?

```python
query_2 = "SELECT COUNT(*) FROM Fires WHERE FIRE_CODE IS NULL;"

df_2 = pd.read_sql_query(query_2, conn)

df_2.head()
```

In the majority of the observations, FIRE_CODE is null, meaning that it might not make sense to include it in our further work.

## How many observations have NULL values at the state variable?

```python
query_3 = "SELECT COUNT(*) FROM Fires WHERE STATE IS NULL;"

df_3 = pd.read_sql_query(query_3, conn)

df_3.head()
```

Wow, all observations contain a STATE value! Nice

## How many observations have NULL values at the county variable?

```python
query_4 = "SELECT COUNT(*) FROM Fires WHERE COUNTY IS NULL;"

df_4 = pd.read_sql_query(query_4, conn)

df_4.head()
```

Roughly one third of the observations don't have COUNTY value. We should discuss it if it'd make sense to filter all these out or ditch the idea of working with data on county level.

## How many different causes are there (STAT_CAUSE_DESCR)?

```python
query_5 = "SELECT STAT_CAUSE_DESCR, COUNT(*) FROM Fires GROUP BY STAT_CAUSE_DESCR;"

df_5 = pd.read_sql_query(query_5, conn)

df_5
```

## How many observations do we have in each state per year?

```python
query_6 = "SELECT FIRE_YEAR, STATE, COUNT(*) FROM Fires GROUP BY FIRE_YEAR, STATE;"

df_6 = pd.read_sql_query(query_6, conn)

df_6
```

## What are the minimum and maximum values in this interval?

```python
df_6.nsmallest(10, 'COUNT(*)')
```

```python
df_6.nlargest(20, 'COUNT(*)')
```
